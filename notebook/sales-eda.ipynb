Null cells in ADDRESLINE2, STATE,POSTALCODE, TERRITORY to be replaced by a simple empty string as it can be just empty, not always used field in forms

```python
sales_table['ADDRESSLINE2'].fillna('', inplace=True)
sales_table['STATE'].fillna('', inplace=True)
sales_table['POSTALCODE'].fillna('', inplace=True)
sales_table['TERRITORY'].fillna('', inplace=True)
```

Check null count after cleaning

```python
sales_table.info()
```

Null cells filled

for dtypes need update:

Numerical: ORDERNUMBER, QUANTITYORDERED, ORDERLINENUMBER, QTR_ID, MONTH_ID, YEAR_ID, MSRP

datetime: ORDERDATE

Converting ORDERDATE to datetime dtype

```python
sales_table['ORDERDATE'] = pd.to_datetime(sales_table['ORDERDATE'])
```

Checking ORDERDATE dtype after conversion

```python
sales_table.dtypes
```

Checking that YEAR_ID match date in ORDERDATE

```python
(sales_table['ORDERDATE'].dt.year == sales_table['YEAR_ID']).value_counts()
```

Checking that MONTH_ID match date in ORDERDATE

```python
(sales_table['ORDERDATE'].dt.month == sales_table['MONTH_ID']).value_counts()
```

MONTH_ID match ORDERDATE

Checking that QTR_ID match date in ORDERDATE, through chekcs with MONTH_ID as it matches ORDERDATE

```python
(sales_table['QTR_ID'] == (sales_table['MONTH_ID'] - 1) // 3 + 1).value_counts()
```

QTR_ID match ORDERDATE

Checking YEARID range

Optimising YEAR_ID, MONTH_ID, QTR_ID ORDERNUMBER, QUANTITYORDERED, ORDERLINENUMBER, MSRP dtypes based on their min, max values

```python
sales_table['YEAR_ID'] = sales_table['YEAR_ID'].astype(np.uint16)
sales_table['MONTH_ID'] = sales_table['MONTH_ID'].astype(np.uint8)
sales_table['QTR_ID'] = sales_table['QTR_ID'].astype(np.uint8)
sales_table['ORDERNUMBER'] = sales_table['ORDERNUMBER'].astype(np.uint16)
sales_table['QUANTITYORDERED'] = sales_table['QUANTITYORDERED'].astype(np.uint8)
sales_table['ORDERLINENUMBER'] = sales_table['ORDERLINENUMBER'].astype(np.uint8)
sales_table['MSRP'] = sales_table['MSRP'].astype(np.uint8)
```
